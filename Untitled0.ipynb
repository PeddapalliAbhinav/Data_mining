```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('/content/titanic_clean.csv')
```

```python
summary = data.describe()
print(summary)
print(data.describe(include=['object']))
```

```python
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
data['pclass'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('age')
plt.show()
```

```python
sns.pairplot(data[['age', 'fare', 'sibsp', 'parch']].dropna())
plt.show()
```

```python
sns.countplot(x='survived', data=data, palette='Set2',hue='sex')
plt.legend(loc='upper right')
plt.title('survival count')
plt.xlabel('survived')
plt.ylabel('count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data, palette='muted', hue='sex')
plt.legend()
plt.title('Age Distribution by Passenger Class')
plt.show()
```
